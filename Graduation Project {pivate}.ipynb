# 3- Deteching With Outliers

```python
Num_Coulmns = x.select_dtypes(include = ["float64"]).columns
for col in Num_Coulmns:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=x[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
for i,col in enumerate(Num_Coulmns):
    Q1 = x[col].quantile(0.25)
    Q3 = x[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = x[(x[col] < lower_bound) | (x[col] > upper_bound)][col]
    print(f"{i + 1}- {col}: {len(outliers)} outliers")
```

# 10- Encoding

```python
categorical_cols = ['Gender', 'Exercise Habits', 'Smoking', 'Family Heart Disease', 'Diabetes',
                    'High Blood Pressure', 'Low HDL Cholesterol', 'High LDL Cholesterol',
                    'Alcohol Consumption', 'Stress Level', 'Sugar Consumption', 'Heart Disease Status']

for i , col in enumerate(categorical_cols):
    print(f"{ i + 1 }- The Values In This Coulmn Is: {x[col].unique()}")
```

```python
x_encoded = x.copy()

binary_columns = ['Gender', 'Family Heart Disease', 'Diabetes', 'High Blood Pressure',
                  'Low HDL Cholesterol', 'High LDL Cholesterol', 'Heart Disease Status','Smoking']

ordinal_columns = ['Exercise Habits', 'Stress Level','Alcohol Consumption','Sugar Consumption']

z = {'Low': 0, 'Medium': 1, 'High': 2}

for col in binary_columns:
    print(f"{col} Before mapping:")
    print(x_encoded[col].value_counts(dropna=False))
    x_encoded[col] = x_encoded[col].map({'No': 0, 'Yes': 1, 'Male': 0, 'Female': 1})
    print(f"{col} After mapping:")
    print(x_encoded[col].value_counts(dropna=False))
    print("="*40)

for col in ordinal_columns:
    print(f"{col} Before mapping:")
    print(x_encoded[col].value_counts(dropna=False))
    x_encoded[col] = x_encoded[col].map(z)
    print(f"{col} After mapping:")
    print(x_encoded[col].value_counts(dropna=False))
    print("="*40)
```

# 11- Scaling

```python
from sklearn.preprocessing import StandardScaler

numeric_columns = x_encoded.select_dtypes(include=['float64', 'int64']).columns.tolist()

exclude_cols = binary_columns + ordinal_columns

cols_to_scale = [col for col in numeric_columns if col not in exclude_cols]

scaler = StandardScaler()
x_encoded[cols_to_scale] = scaler.fit_transform(x_encoded[cols_to_scale])

print(x_encoded.head())
```

# 12- Feature Engeneering

```python
# Add BMI_Category Coulmn:
#===========================
def bmi_category(bmi):
    if bmi < 18.5:
        return "Underweight"
    elif bmi < 25:
        return "Normal weight"
    elif bmi < 30:
        return "Overweight"
    elif bmi < 35:
        return "Obesity Class I"
    elif bmi < 40:
        return "Obesity Class II"
    else:
        return "Obesity Class III"

x["BMI Category"] = x["BMI"].apply(bmi_category)
```

```python
x.shape
```

```python
x.info()
```

```python
# Add Age_Group Coulmn
#=============================
def age_group(age):
    if 18 <= age <= 30:
        return 'Young'
    elif 31 <= age <= 50:
        return 'Middle-aged'
    elif age >= 51:
        return 'Senior'
    else:
        return 'Unknown'  

x['Age Group'] = x['Age'].apply(age_group)
```

```python
x.info()
```

```python
# Add Sleep_Category Coulmn
#=============================
def sleep_category(hours):
    if hours < 4:
        return 'Very Short'
    elif 4 <= hours < 6:
        return 'Short'
    elif 6 <= hours <= 8:
        return 'Normal'
    else:
        return 'Long'

x['Sleep Category'] = x['Sleep Hours'].apply(sleep_category)
```

# 13- Spliting And Smote In Target Coulmn

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split

X = x_encoded.drop("Heart Disease Status", axis=1)
y = x_encoded["Heart Disease Status"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

smote = SMOTE(random_state=42)
X_train_bal, y_train_bal = smote.fit_resample(X_train, y_train)

print("Before SMOTE:", y_train.value_counts())
print("After SMOTE:", y_train_bal.value_counts())
# 6400 ===> 8000 * 80%
# 1600 ===> 2000 * 80% 
```

# 1- Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Build Model
logreg = LogisticRegression(max_iter=2000, class_weight="balanced", random_state=42)
logreg.fit(X_train_bal, y_train_bal)

# Prediction
y_pred = logreg.predict(X_test)

# Accuracy
acc = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)

print(f"Logistic Regression Accuracy:, {acc * 100}%")
print("==="*49)
print("Logistic Regression Confusion Matrix:\n", cm)
cm_LR = confusion_matrix(y_test,y_pred)
ConfusionMatrixDisplay(cm_LR).plot(cmap='plasma')
```

# 2- RandomForest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Build Model
rf = RandomForestClassifier(class_weight="balanced", random_state=42)
rf.fit(X_train_bal, y_train_bal)

# Prediction
y_pred = rf.predict(X_test)

# Accuracy
acc = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)

print(f"Logistic Regression Accuracy:, {acc * 100}%")
print("==="*49)
print("RandomForest Confusion Matrix:\n", cm)
cm_LR = confusion_matrix(y_test,y_pred)
ConfusionMatrixDisplay(cm_LR).plot(cmap='plasma')
```

# 3- Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
# بناء الموديل
nb = GaussianNB()
nb.fit(X_train_bal, y_train_bal)

# التنبؤ
y_pred = nb.predict(X_test)

# التقييم
acc = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)

print(f"Logistic Regression Accuracy:, {acc * 100}%")
print("==="*49)
print("Naive Bayes Confusion Matrix:\n", cm)
cm_LR = confusion_matrix(y_test,y_pred)
ConfusionMatrixDisplay(cm_LR).plot(cmap='plasma')
```

# 4- Comparision Of All Accuracy Models

```python
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

accuracy_scores = {
    "Logistic Regression": accuracy_score(y_test, logreg.predict(X_test)),
    "RandomForest": accuracy_score(y_test, rf.predict(X_test)),
    "Naive Bayes": accuracy_score(y_test, nb.predict(X_test))
}

models = list(accuracy_scores.keys())
scores = list(accuracy_scores.values())

plt.figure(figsize=(8,5))
plt.bar(models, scores,color = ["navy", "darkred", "forestgreen"])
plt.ylim(0,1)                                                           # Define Range Of Accuracy Bar 0 ==> 1 
plt.ylabel("Accuracy")
plt.title("Comparison of Accuracy Between Models")
for i, v in enumerate(scores):
    plt.text(i, v + 0.02, f"{v:.2f}", ha='center', fontweight='bold')
plt.show()
# v + 0.02 ===> Write A Number On A Surface Of Bar
```
